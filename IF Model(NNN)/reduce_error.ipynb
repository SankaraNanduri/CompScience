```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
import numpy as np

# Load training data
data = pd.read_csv('updated_membrane_potentials.csv')
features = data[['Neuron1_V_mV', 'Neuron2_V_mV', 'Neuron3_V_mV']]
target = data['Neuron4_V_cal']  # Ensure this column name matches your dataset

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create a pipeline for polynomial regression
pipeline = Pipeline([
    ('poly_features', PolynomialFeatures(degree=2)),
    ('scaler', StandardScaler()),
    ('lin_reg', LinearRegression())
])

# Fit the model
pipeline.fit(X_train, y_train)

# Predict on the test set
y_pred = pipeline.predict(X_test)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Chi-Square Error
chi_square_error = np.sum(((y_test - y_pred) ** 2) / y_pred)
print(f'Chi-Square Error: {chi_square_error}')

# MAPE
mape = np.mean(np.abs((y_test - y_pred) / (y_test + 1e-10))) * 100
print(f'Mean Absolute Percentage Error: {mape}%')

# Load new data from a CSV file for prediction
new_data_df = pd.read_csv('updated_membrane_potentials.csv')
new_data = new_data_df[['Neuron1_V_mV', 'Neuron2_V_mV', 'Neuron3_V_mV']]

# Predict new activation values using the loaded new data
new_predictions = pipeline.predict(new_data)
print(f'Predicted new activations for neuron4: {new_predictions}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline

# Load training data
data = pd.read_csv('updated_membrane_potentials.csv')
features = data[['Neuron1_V_mV', 'Neuron2_V_mV', 'Neuron3_V_mV']]
target = data['Neuron4_V_cal']  # Ensure this column name matches your dataset

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create a pipeline for polynomial regression
pipeline = Pipeline([
    ('poly_features', PolynomialFeatures(degree=2)),
    ('scaler', StandardScaler()),
    ('lin_reg', LinearRegression())
])

# Fit the model
pipeline.fit(X_train, y_train)

# Predict on the test set
y_pred = pipeline.predict(X_test)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Chi-Square Error
chi_square_error = np.sum(((y_test - y_pred) ** 2) / y_pred)
print(f'Chi-Square Error: {chi_square_error}')

# MAPE
mape = np.mean(np.abs((y_test - y_pred) / (y_test + 1e-10))) * 100
print(f'Mean Absolute Percentage Error: {mape}%')

# Load new data from a CSV file for prediction
new_data_df = pd.read_csv('updated_membrane_potentials.csv')
new_data = new_data_df[['Neuron1_V_mV', 'Neuron2_V_mV', 'Neuron3_V_mV']]

# Predict new activation values using the loaded new data
new_predictions = pipeline.predict(new_data)

# Print new predictions
print(f'Predicted new activations for neuron4: {new_predictions}')

# Plot actual vs. predicted values for new data
plt.figure(figsize=(10, 6))
plt.scatter(new_data_df.index, new_data_df['Neuron4_V_cal'], color='blue', label='Actual', alpha=0.6)
plt.scatter(new_data_df.index, new_predictions, color='red', label='Predicted', alpha=0.6)
plt.title('Comparison of Actual and Predicted Neuron4 Activations')
plt.xlabel('Data Index')
plt.ylabel('Neuron4 Activation')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline

# Load training data
data = pd.read_csv('updated_membrane_potentials.csv')
features = data[['Neuron1_V_mV', 'Neuron2_V_mV', 'Neuron3_V_mV']]
target = data['Neuron4_V_cal']  # Ensure this column name matches your dataset

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create a pipeline for polynomial regression
pipeline = Pipeline([
    ('poly_features', PolynomialFeatures(degree=2)),
    ('scaler', StandardScaler()),
    ('lin_reg', LinearRegression())
])

# Fit the model
pipeline.fit(X_train, y_train)

# Predict on the test set
y_pred = pipeline.predict(X_test)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Chi-Square Error
chi_square_error = np.sum(((y_test - y_pred) ** 2) / y_pred)
print(f'Chi-Square Error: {chi_square_error}')

# MAPE
mape = np.mean(np.abs((y_test - y_pred) / (y_test + 1e-10))) * 100
print(f'Mean Absolute Percentage Error: {mape}%')

# Load new data from a CSV file for prediction
new_data_df = pd.read_csv('updated_membrane_potentials.csv')
new_data = new_data_df[['Neuron1_V_mV', 'Neuron2_V_mV', 'Neuron3_V_mV']]

# Predict new activation values using the loaded new data
new_predictions = pipeline.predict(new_data)

# Add predictions to the dataframe
new_data_df['Neuron4_V_pred'] = new_predictions

# Print new predictions
#print(f'Predicted new activations for neuron4: {new_predictions}')

# Save the new dataframe with predictions to a new CSV file
new_data_df.to_csv('updated_membrane_potentials.csv', index=False)

# Plot actual vs. predicted values for new data
plt.figure(figsize=(10, 6))
plt.scatter(new_data_df.index, new_data_df['Neuron4_V_cal'], color='blue', label='Actual', alpha=0.6)
plt.scatter(new_data_df.index, new_data_df['Neuron4_V_pred'], color='red', label='Predicted', alpha=0.6)
plt.title('Comparison of Actual and Predicted Neuron4 Activations')
plt.xlabel('Data Index')
plt.ylabel('Neuron4 Activation')
plt.legend()
plt.show()
```
